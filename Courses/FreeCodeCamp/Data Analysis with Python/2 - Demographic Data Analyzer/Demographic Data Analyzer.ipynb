```python
df.groupby('education').size()
```

```python
h_e = df[df.education.isin(['Bachelors', 'Masters', 'Doctorate'])]
l_e = df[~df.education.isin(['Bachelors', 'Masters', 'Doctorate'])]
```

```python
h_e_r = h_e[h_e['salary']=='>50K']
l_e_r = l_e[l_e['salary']=='>50K']
```

```python
print(h_e.shape[0])
print(h_e_r.shape[0])
print(h_e_r.shape[0]/h_e.shape[0])
```

```python
print(l_e.shape[0])
print(l_e_r.shape[0])
print(l_e_r.shape[0]/l_e.shape[0])
```

```python
min_hrs = df['hours-per-week'].min()
```

```python
min_workers = df.loc[(df['hours-per-week']==min_hrs)]
min_workers.shape[0]
```

```python
min_workers[min_workers['salary']=='>50K'].shape[0]/min_workers.shape[0]*100
```
