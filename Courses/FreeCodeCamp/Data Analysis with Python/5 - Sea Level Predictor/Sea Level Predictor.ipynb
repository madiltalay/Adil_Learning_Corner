```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
def draw_plot():
    # Read data from file
    df = pd.read_csv('epa-sea-level.csv')
    fig, ax = plt.subplots(3,1, figsize=(9,18))

    # Create scatter plot
    x1 = df['Year']
    y1 = df['CSIRO Adjusted Sea Level']
    ax[0].scatter(x1,y1)

    # Create first line of best fit
    slope_1, intercept_1, r_1, p_1, se_1 = linregress(x1, y1)
    xh1 = np.arange(1880, 2051, 10)
    yh1 = intercept_1+slope_1*xh1

    ax[1].scatter(x1,y1)
    ax[1].plot(xh1,yh1)

    # Create second line of best fit
    df_new = df[df['Year']>=2000]
    x2 = df_new['Year']
    y2 = df_new['CSIRO Adjusted Sea Level']
    slope_2, intercept_2, r_2, p_2, se_2 = linregress(x2, y2)
    xh2 = np.arange(2000, 2051, 10)
    yh2 = intercept_2+slope_2*xh2

    ax[2].scatter(x2,y2)
    ax[2].plot(xh2,yh2)

    # Add labels and title
    ax[0].set_xlabel('Year')
    ax[0].set_ylabel('Sea Level (inches)')
    ax[0].set_title('Rise in Sea Level')

    ax[1].set_xlabel('Year')
    ax[1].set_ylabel('Sea Level (inches)')
    ax[1].set_title('Rise in Sea Level')

    ax[2].set_xlabel('Year')
    ax[2].set_ylabel('Sea Level (inches)')
    ax[2].set_title('Rise in Sea Level')
    
    # Save plot and return data for testing (DO NOT MODIFY)
    plt.savefig('sea_level_plot.png')
    return plt.gca()
```

```python
ax = draw_plot()
```
